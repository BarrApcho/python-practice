```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

print("Libraries imported successfully.")
```

**Razonamiento**:
Crearé una lista de categorías de Python, luego inicializaré un diccionario vacío y lo llenaré con datos numéricos generados aleatoriamente, datos categóricos elegidos de la lista y datos secuenciales, asegurándome de que todas las columnas tengan la misma longitud.






```python
import numpy as np

# 1. Cree una lista de cadenas de Python para datos categóricos
categories = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']

# Defina una longitud consistente para todas las columnas de datos
data_length = 100

# 2. Inicializar un diccionario vacío
data = {}

# 3. Agregue una clave para datos numéricos usando np.aleatorio.randint()
data['snacks'] = np.random.randint(1, 101, data_length)

# 4. Agregue una clave para datos categóricos usando np.aleatorio.elección()
data['Datos categóricos'] = np.random.choice(categories, data_length)

# 5. Agregue una clave final para datos secuenciales usando np.arange()
data['Id_secuenciAL'] = np.arange(1, data_length + 1)

# 6. Imprima el diccionario creado para inspeccionar los datos generados
print("Diccionario de Datos Generados:")
```

**Razonamiento**:
El paso anterior generó con éxito diversos datos en un formato de diccionario. Ahora, convertiré este diccionario en un marco de datos de pandas, según lo requiera la tarea principal, y mostraré su encabezado para verificar la estructura y el contenido.


```python
# Convierte el diccionario en un marco de datos de pandas
df = pd.DataFrame(data)

# Muestra las primeras filas del marco de datos
print("Encabezado del marco de datos generado:")
print(df.head())
```

**Razonamiento**:
Crearé un diagrama de líneas



```python
# Crea el trazado de líneas
plt.figure(figsize=(10, 6))
plt.plot(df['Id_secuenciAL'], df['snacks'], marker='o', linestyle='-', color='skyblue', label='Numerical Data Trend')

# Agregar título y etiquetas
plt.title('Gráfico Lineal de snacks Sobre IDSecuencia')
plt.xlabel('ID secuencial')
plt.ylabel('Snacks')

# Agregar leyenda y cuadrícula
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Razonamiento**:
Siguiendo la explicación en la celda de rebajas anterior, ahora crearé un gráfico de columnas (barras)



```python

# Calcula la frecuencia de cada categoría
category_counts = df['Datos categóricos'].value_counts()

# Crea el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values, color='lightcoral')

# Agregar título y etiquetas
plt.title('Gráfico de Columnas de cartegorias por snacks')
plt.xlabel('Categoría')
plt.ylabel('snacks')

# Gire las etiquetas del eje x si es necesario para facilitar la lectura
plt.xticks(rotation=45, ha='right')

# Agregar cuadrícula para una mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

**Razonamiento**:
Siguiendo la explicación en la celda de rebajas anterior, ahora crearé un gráfico circular usando


```python
# Suponiendo que category_counts se calculó en el paso anterior
# category_counts = df ['Categorical_Data'].valor_cuentas()

# Crea el gráfico circular
plt.figure(figsize=(8, 8))
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Añadir título
plt.title('Gráfico Circular de Distribución de snacks por las categorias')

# Agregue una leyenda fuera del círculo para una mejor legibilidad
plt.legend(category_counts.index, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Asegúrate de que el círculo se dibuje proporcionalmente
plt.axis('equal')


plt.tight_layout()
plt.show()
```

**Razonamiento**:
Siguiendo la explicación en la celda de rebajas anterior, ahora crearé un gráfico de área usando

```python

# Crea el diagrama de área
fig = px.area(df, x='Id_secuenciAL', y='snacks', title='Gráfica de Área de snacks Sobre Secuencia')

# Actualice el diseño para una mejor legibilidad (opcional, px a menudo maneja esto bien)
fig.update_layout(
    xaxis_title='ID secuencial',
    yaxis_title='snacks',
    hovermode='x unified'
)


fig.show()
```

**Razonamiento**:
Siguiendo la explicación en la celda de rebajas anterior, ahora crearé un diagrama de caja



```python
# Crea el diagrama de caja
fig = px.box(df, x='Datos categóricos', y='snacks', title='Diagrama de snacks por Categoría', color='Datos categóricos')

# Actualizar el diseño para una mejor legibilidad
fig.update_layout(
    xaxis_title='Categoría',
    yaxis_title='cantidad de snacks',
    legend_title='Categories'
)


fig.show()
```

**PREGUNTA 2**

Realiza el gráfico de las ecuaciones y=x-2 , y=6-x , y^2-6x=0, coloca su titulo , leyenda ,  puntos
de intersección para las rectas.

**Razonamiento**:
Definiré los rangos x y calcularé los valores y correspondientes para cada una de las tres ecuaciones dadas ` ' y = x-2`, `y=6-x` e `y^2-6x=0` (o `y = +/- sqrt (6x)`). Luego, trazaré todas las ecuaciones usando Matplotlib, asegurándome de que los rangos adecuados para x e y muestren las curvas claramente, como se especifica en las instrucciones.



```python
# Define un rango de valores x
x_single = np.linspace(-10, 10, 400)

# Calcular valores y para y = x - 2
y= x_single - 2

# Crea la trama
plt.figure(figsize=(10, 8))
plt.plot(x_single, y, label='y = x - 2', color='purple')

# Agregar título y etiquetas
plt.title('Gráfico de la Ecuación: y = x - 2')
plt.xlabel('x')
plt.ylabel('y')

# Agregar leyenda y cuadrícula
plt.legend()
plt.grid(True)

# Agregue ejes en el origen para mayor claridad
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python


# Define un rango de valores x
x_single = np.linspace(-10, 10, 400)

# Calcular valores y para y = 6-x
y_single = 6 - x_single

# Crea la trama
plt.figure(figsize=(10, 8))
plt.plot(x_single, y_single, label='y = 6 - x', color='orange')

# Agregar título y etiquetas
plt.title('Gráfico de la Ecuación: y = 6 - x')
plt.xlabel('x')
plt.ylabel('y')

# Agregar leyenda y cuadrícula
plt.legend()
plt.grid(True)

# Agregue ejes en el origen para mayor claridad
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python


# Defina un rango de valores x para la parábola. x no puede ser negativo.
x_parabola_single = np.linspace(0, 10, 400)

# Calcular valores y para y = + / - sqrt ( 6x)
y_parabola_positive = np.sqrt(6 * x_parabola_single)
y_parabola_negative = -np.sqrt(6 * x_parabola_single)

# Crea la trama
plt.figure(figsize=(10, 8))
plt.plot(x_parabola_single, y_parabola_positive, label='y^2 = 6x (y >= 0)', color='green')
plt.plot(x_parabola_single, y_parabola_negative, label='y^2 = 6x (y < 0)', color='darkgreen', linestyle='--')

# Agregar título y etiquetas
plt.title('Gráfico de la Ecuación: y^2 - 6x = 0')
plt.xlabel('x')
plt.ylabel('y')

# Agregar leyenda y cuadrícula
plt.legend()
plt.grid(True)

# Agregue ejes en el origen para mayor claridad
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)


plt.tight_layout()
plt.show()
```

```python


# 1. Defina un rango de valores x para las ecuaciones lineales
x = np.linspace(-5, 10, 400)

# 2. Calcule los valores y para la primera ecuación lineal
y1 = x - 2

# 3. Calcule los valores y para la segunda ecuación lineal
y2 = 6 - x

# 4. Para la ecuación parabólica y^2 - 6x = 0 (o y = +/- sqrt(6x)), defina un rango separado de valores de x
x_parabola = np.linspace(0, 10, 400) # x no puede ser negativo para sqrt(6x)

# 5. Calcule las dos ramas de los valores y para la parábola
y3_positive = np.sqrt(6 * x_parabola)
y3_negative = -np.sqrt(6 * x_parabola)

# Calcula el punto de intersección de las dos ecuaciones lineales:
# x - 2 = 6 - x
# 2x = 8
# x = 4
# y = 4 - 2 = 2
intersection_x = 4
intersection_y = 2

# 6. Crea una figura con un tamaño apropiado
plt.figure(figsize=(10, 8))

# 7. Traza la primera línea
plt.plot(x, y1, label='y = x - 2', color='blue')

# 8. Traza la segunda línea
plt.plot(x, y2, label='y = 6 - x', color='red')

# 9. Traza ambas ramas de la parábola
plt.plot(x_parabola, y3_positive, label='y^2 = 6x', color='green')
plt.plot(x_parabola, y3_negative, color='green', linestyle='--') # Sin etiqueta para la segunda rama, ya que es parte de la misma ecuación

# Traza el punto de intersección de las dos ecuaciones lineales
plt.plot(intersection_x, intersection_y, 'ko', markersize=8, label=f'Intersección: ({intersection_x}, {intersection_y})')

# Agregar título, etiquetas y leyenda
plt.title('Gráfico de Ecuaciones: y=x-2, y=6-x, y^2=6x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-10, 10) # Establezca límites en el eje y para una mejor visualización
plt.xlim(-5, 10) # Establecer límites del eje x para una mejor visualización


plt.tight_layout()
plt.show()
```
